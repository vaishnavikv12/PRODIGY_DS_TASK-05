## Analyze traffic accident data to identify patterns related to road conditions, weather, and time of day. Visualize accident hotspots and contributing factors.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('TASK 05.csv')
data
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
data.dropna(inplace=True)
```

```python
data['Age_band_of_driver'] = data['Age_band_of_driver'].astype('category')
data['Driving_experience'] = data['Driving_experience'].astype('category')
```

```python
severity_counts = data['Accident_severity'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Accident_severity', palette='viridis')
plt.title('Accident Severity Distribution')
plt.xlabel('Accident Severity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Road_surface_type', hue='Accident_severity', palette='coolwarm')
plt.title('Accident Count by Road Surface Type')
plt.xlabel('Road Surface Type')
plt.ylabel('Count')
plt.legend(title='Accident Severity')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Light_conditions', hue='Accident_severity', palette='Set2')
plt.title('Accident Count by Light Conditions')
plt.xlabel('Light Conditions')
plt.ylabel('Count')
plt.legend(title='Accident Severity')
plt.xticks(rotation=45)
plt.show()
```
